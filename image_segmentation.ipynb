```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = cv2.imread('./image1.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
def colorrecocognized(centers):
  
    for each_col in centers:
        img = np.zeros((10,10,3),dtype="uint8")
        img[:,:,:] = each_col
        plt.figure()
        plt.imshow(img)
        
           
```

```python
def imagesegmentation(im,n_color= 5):
    km = KMeans(n_clusters=n_color)
    org_shape= im.shape
    pix = im.reshape((-1,org_shape[2]))
    km.fit(pix)
    centers = km.cluster_centers_
      
    labels = km.labels_
    
    for ix in range(pix.shape[0]):
        pix[ix,:] = centers[labels[ix]]
    
    final_image = pix.reshape(org_shape)
    
    
    return final_image,centers  
```

```python
finalimg1,_ = imagesegmentation(img,20)
plt.imshow(finalimg1)
```

```python
finalimg2,_ = imagesegmentation(img,10)
plt.imshow(finalimg2)
```

```python
finalimg3,centers = imagesegmentation(img,5)
plt.imshow(finalimg3)
```

```python
colorrecocognized(centers)
```
